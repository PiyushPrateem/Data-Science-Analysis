# 1.	Read the data and remove all null or empty values.

```python
from csv import reader

# Initialize an empty list to store transactions
groceries = []

# Read the CSV file and filter out empty strings
with open('Dataset_Day15.csv', 'r') as csvfile:
    read = reader(csvfile)
    for row in read:
        # Filter out empty strings
        row_cleaned = [item for item in row if item]
        groceries.append(row_cleaned)
```

```python
groceries[0:3]
```

### 2.	Transform the data using TransactionEncoder to perform Market Basket Analysis.

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()

transactions = encoder.fit(groceries).transform(groceries)
transactions
import pandas as pd

itemsets = pd.DataFrame(transactions, columns = encoder.columns_)
```

```python
display(itemsets.head())
display(itemsets.info())
```

### 3.	 Use min_support = 0.02 to find frequent itemsets.

```python
from mlxtend.frequent_patterns import apriori
# Find frequent itemsets with a minimum support of 0.02
frequent_itemsets = apriori(itemsets, min_support = 0.02, use_colnames = True)
frequent_itemsets
```

```python
# Sort the frequent itemsets by support in descending order
frequent_itemsets.sort_values('support', ascending = False)
```

### 4.	Use the frequent itemsets to create association rules (take min_threshold as 15%) and evaluate the rules for the following metrics

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.15)
rules
```

```python
rules.describe()
```
