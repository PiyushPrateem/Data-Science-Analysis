```python
#Find all unique values of columns: “name”, “mfr”, “vitamins” and store them in separate numpy arrays and print them.

import csv 
import numpy as np
import pandas as pd

#csv file into dataframe
with open('Dataset_Day2.csv','r') as f:
    df=pd.read_csv(f)

#unique values for columns
u_name=df['name'].unique()
u_mfr=df['mfr'].unique()
u_vit=df['vitamins'].unique()

#store these into arrays
u_arr=np.array(u_name)
m_arr=np.array(u_mfr)
v_arr=np.array(u_vit)

#print the arrays
print("Unique values 'name' arrays:",u_arr)
print("Unique values 'mfr' arrays:",u_mfr)
print("Unique values 'vitamins' arrays:",u_vit)
```

```python
#Create a new dataframe with all columns for which,‘sodium’ is greater than 100 AND ‘protein’ is less than 3
#Name this dataframe:  df_HighSodLowProt
#Print this dataframe.

#all columns for which,‘sodium’ is greater than 100 AND ‘protein’ is less than 3
df_HighSodLowProt=df[(df['sodium'] > 100) & (df['protein'] < 3)]

#print the dataframe
print(df_HighSodLowProt)
```

```python
#From the dataframe in 2. print the average ‘calories’ by ‘mfr’  and also print the ‘mfr’ with the highest average ‘calories’.

new_df=df_HighSodLowProt[['mfr','calories']]
avg_cal=new_df.groupby(['mfr']).mean(['calories']).reset_index()
display(avg_cal)

max_val=avg_cal['calories'].max()
avg_cal[avg_cal['calories'] == max_val]
```
